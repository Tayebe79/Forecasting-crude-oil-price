```python
import pandas as pd
from prophet import Prophet
import pandas as pd
import numpy as np
from sklearn.linear_model import LinearRegression, Ridge, Lasso
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_squared_error, r2_score, accuracy_score, mean_absolute_error
```

```python
df = pd.read_excel('crude oil dataset/Brent 6monthly2017prphet.xlsx')
```

```python
x = np.array(df.y)
y = df.ds
x = np.array(x)
x1 = x

#specify the number of tests-1 here
num = 2

n_test = len(x) - num - 5

xtrain = x[:n_test]
ytrain = y[:n_test]
#test labels should be given to the model in the following array
a = [98.83,106.92,111.156666666667]
ytest = a
ytest = np.array(ytest)



ytrain = np.array(ytrain)

xtrain = np.array(xtrain)
```

```python
mae = []
rmse = []
y_pred1 = []
for i in range(0,int((len(x)-num-5)/3)-1):
    for j in range(0,3):
        xtrain = np.delete(xtrain, 0)
        ytrain = np.delete(ytrain, 0)
    df = pd.DataFrame(ytrain,columns=["ds"])
    df.insert(loc=1,
          column='y',
          value=xtrain)
    m = Prophet()
    m.fit(df)
    future = m.make_future_dataframe(periods=7)
    future.tail()
    forecast = m.predict(future)
    fig1 = m.plot(forecast)
    forecast[['ds', 'yhat', 'yhat_lower', 'yhat_upper']].tail()
    c = np.array(forecast.yhat)
    b = []
    b.append(c[len(c)-3])
    b.append(c[len(c)-2])
    b.append(c[len(c)-1])
    print(b)
    mae.append(mean_absolute_error(ytest,b))
    rmse.append(np.sqrt(mean_squared_error(ytest,b)))
    print("the ith time: ",i,"\nMAE is: ",format(mean_absolute_error(ytest,b),".3f"))
    print("RMSE is: ",format(np.sqrt(mean_squared_error(ytest,b)),".3f"))
    
```

```python
import matplotlib.pyplot as plt
x_axis = list(range(0,int((len(x)-num-5)/3)-1))
y_axis = rmse
plt.plot(x_axis,y_axis)
plt.show()
```
